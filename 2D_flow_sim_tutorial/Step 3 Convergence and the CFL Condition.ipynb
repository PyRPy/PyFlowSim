```python
import numpy
from matplotlib import pyplot
%matplotlib inline
```

```python
def linearconv(nx):
    dx = 2 / (nx - 1)
    nt = 20
    dt = 0.025
    c = 1
    
    u = numpy.ones(nx)
    u[int(0.5/dx): int(1/dx + 1)] = 2
    un = numpy.ones(nx) # placeholder arrray
    
    for n in range(nt):
        un = u.copy()
        for i in range(1, nx):
            u[i] = un[i] - c * dt/dx * (un[i] - un[i-1])
    pyplot.plot(numpy.linspace(0, 2, nx), u)
```
